```python
# Data Load 
import pandas as pd
import os, IPython
report_files_path = 'ml_perf_report/'
report_files = os.listdir(report_files_path)

master_df = []
master_df2 = []
for r_file in report_files:
   repo_name = os.path.splitext(r_file)[0]
   repo_df = pd.read_csv (report_files_path+r_file)
   #print(repo_df.set_index('classifier').T)
   repo_df['Project'] = repo_name
   master_df.append(repo_df)
   master_df2.append(repo_df.set_index('classifier').T)
```
