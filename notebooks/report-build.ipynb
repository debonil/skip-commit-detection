```python
import pandas as pd
import os, IPython
report_files_path = '../ml_perf_report/'

repo_df = pd.read_csv (report_files_path+'final_merged.csv',)
```

```python
repo_df = repo_df.drop('Unnamed: 0',axis=1)
```

```python
repo_df
```

```python
def matric_report(summary_df,matric='accuracy_score',top=10):
    result = summary_df.pivot_table(matric, ['classifier'], 'Project')\
            .sort_values(by=['eBay_parallec','jMotif_GI','tracee_contextlogger','jMotif_SAX'],ascending=False)\
                .sort_values(by=['Neural Net','Random Forest','RBF SVM','Logistic Regression','Linear SVM',],axis=1,ascending=False).transpose().head(5)
    result.loc['Average'] = result.mean()
    result.loc['Median'] = result.median()
    result.to_csv(f'{report_files_path}matric_wise_summary/{matric}_top_{top}.csv')
    result.to_html(f'{report_files_path}matric_wise_summary/{matric}_top_{top}.html')
    result.to_markdown(f'{report_files_path}matric_wise_summary/{matric}_top_{top}.txt')
    return result

IPython.display.display(matric_report(repo_df,'accuracy_score',5))
IPython.display.display(matric_report(repo_df,'f1_score',5))
IPython.display.display(matric_report(repo_df,'precision_score',5))
IPython.display.display(matric_report(repo_df,'recall_score',5))
IPython.display.display(matric_report(repo_df,'roc_auc_score',5))
IPython.display.display(matric_report(repo_df,'accuracy_score'))
IPython.display.display(matric_report(repo_df,'f1_score'))
IPython.display.display(matric_report(repo_df,'precision_score'))
IPython.display.display(matric_report(repo_df,'recall_score'))
IPython.display.display(matric_report(repo_df,'roc_auc_score'))
        
```

```python
score_thresold = 0.75
metric = 'f1_score'
best_projects = repo_df[repo_df[metric] > score_thresold]['Project'].unique()
best_classifier = repo_df[repo_df[metric] > score_thresold]['classifier'].unique()
```

```python
best_projects, best_classifier
```

```python
# Data Load 
import pandas as pd
import os
processed_files_path = 'processed_data'
processed_files = [os.path.join(processed_files_path , f"{f}.csv") for f in best_projects]

print(processed_files)
li = []
for filename in processed_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

commits_df = pd.concat(li, axis=0, ignore_index=True)
```

```python
commits_df = commits_df.loc[:, (commits_df != 0).any(axis=0)].fillna(0.0)
commits_df['LABEL']=commits_df['LABEL'].map(lambda x: (x+1)%2)
print(commits_df['LABEL'].map(lambda x: (x+1)%2).unique())
```

```python
from skip_commit.ml_training import generate_all_clf_perf,prepare_and_split_data,classifiers_names


normed_train_data, training_labels, normed_test_data, testing_labels = prepare_and_split_data(commits_df)
classifiers_perf = generate_all_clf_perf(normed_train_data, training_labels, normed_test_data, testing_labels,
[classifiers_names.index(i) for i in best_classifier])


pd.DataFrame(classifiers_perf).to_csv(
                f'{report_files_path}{metric}_gt{score_thresold*100}.csv', index=False)
```
